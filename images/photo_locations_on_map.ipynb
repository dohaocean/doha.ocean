```python
from exif import Image as ExifImage

def show_exif_data(image_path):
    """Displays selected EXIF metadata of an image."""
    with open(image_path, 'rb') as img_file:
        img = ExifImage(img_file)
        if not img.has_exif:
            print(f"No EXIF metadata found in {image_path}.")
            return

        # Define common EXIF attributes to display
        exif_attributes = [
            'make', 'model', 'datetime', 'gps_latitude', 
            'gps_latitude_ref', 'gps_longitude', 'gps_longitude_ref', 
            'gps_altitude', 'gps_altitude_ref'
        ]
        
        # Only attempt to display attributes that are accessible and meaningful
        for attribute in exif_attributes:
            try:
                value = getattr(img, attribute, None)
                if value is not None:
                    print(f"{attribute}: {value}")
            except (AttributeError, NotImplementedError):
                # Skip attributes that can't be read
                pass

# Example usage
image_path = './IMG_0969.JPG'
show_exif_data(image_path)
```

```python
from PIL import Image
#from exif import Image as ExifImage
import folium
import os
import matplotlib.colors as mcolors
#import matplotlib.cm as cm
import branca.colormap as cm


def get_photo_datetime(image_path):
    """Extracts the date and time from image EXIF data."""
    with open(image_path, 'rb') as img_file:
        img = ExifImage(img_file)
        if not img.has_exif:
            return None, None

        try:
            if hasattr(img, 'datetime'):
                date_time = img.datetime  # e.g., '2024:08:14 10:15:34'
                year = int(date_time.split(':')[0])
                return year, date_time
        except AttributeError:
            return None, None
    return None, None

def get_gps_coordinates(image_path):
    """Extracts GPS coordinates from image EXIF data."""
    with open(image_path, 'rb') as img_file:
        img = ExifImage(img_file)
        if not img.has_exif:
            print(f"No EXIF metadata found in {image_path}.")
            return None
        
        # Define common EXIF attributes to display
        exif_attributes = [
            'make', 'model', 'datetime', 'gps_latitude', 
            'gps_latitude_ref', 'gps_longitude', 'gps_longitude_ref', 
            'gps_altitude', 'gps_altitude_ref'
        ]
        
        # Only attempt to display attributes that are accessible and meaningful
        for attribute in exif_attributes:
            try:
                value = getattr(img, attribute, None)
                if value is not None:
                    print(f"{attribute}: {value}")
                    
                    lat = img.gps_latitude
                    lon = img.gps_longitude
                    lat_ref = img.gps_latitude_ref
                    lon_ref = img.gps_longitude_ref

                    # Convert latitude and longitude to decimal format
                    latitude = convert_to_decimal(lat, lat_ref)
                    longitude = convert_to_decimal(lon, lon_ref)
                    print(latitude)
                    print(longitude)
                    return latitude, longitude
            except (AttributeError, NotImplementedError):
                # Skip attributes that can't be read
                pass

#         #if img.has_exif and 'gps_latitude' in img and 'gps_longitude' in img:
#         if value is not None:
#             lat = img.gps_latitude
#             lon = img.gps_longitude
#             lat_ref = img.gps_latitude_ref
#             lon_ref = img.gps_longitude_ref

#             # Convert latitude and longitude to decimal format
#             latitude = convert_to_decimal(lat, lat_ref)
#             longitude = convert_to_decimal(lon, lon_ref)
#             print(latitude)
#             print(longitude)
#             return latitude, longitude
#         else:
#             print(f"No GPS data found in {image_path}.")
#             return None

def convert_to_decimal(gps_coordinate, gps_ref):
    """Converts GPS coordinates in EXIF format to decimal degrees."""
    decimal = gps_coordinate[0] + gps_coordinate[1] / 60 + gps_coordinate[2] / 3600
    if gps_ref in ['S', 'W']:
        decimal = -decimal
    return decimal

# Set up the base map
map_ = folium.Map(location=[0, 0], zoom_start=2)

# Path to the folder containing photos
#photo_folder = 'path/to/your/photos'
photo_folder = '.'


# Define color scale for years
year_min = 2015  # Start year
year_max = 2019  # End year
#norm = mcolors.Normalize(vmin=year_min, vmax=year_max)
#cmap = cm.get_cmap('viridis')  # Use a colormap (e.g., 'viridis')

colormap = cm.linear.magma.scale(year_min, year_max)
colormap.caption = 'Year the Photo was Taken'


# Loop through each image in the folder
for filename in os.listdir(photo_folder):

    #if filename.lower().endswith(('.jpg', '.jpeg', '.png','.JPG')):
    if filename.lower().endswith(('.jpg', '.jpeg',)):

        image_path = os.path.join(photo_folder, filename)
        print(image_path)
        gps_coords = get_gps_coordinates(image_path)
        
#         # showing only the photo name.
#         if gps_coords:
#             folium.Marker(
#                 location=gps_coords,
#                 popup=filename
#             ).add_to(map_)
#         # showing popup photo directly on map
#         if gps_coords:
#             # Use an HTML popup to display the photo
#             popup_html = f'<img src="{photo_folder}/{filename}" width="300px">'
#             folium.Marker(
#                 location=gps_coords,
#                 popup=folium.Popup(popup_html, max_width=300)
#             ).add_to(map_)
        year, date_time = get_photo_datetime(image_path)
        print(year)

        if gps_coords and year:
#             # Calculate color based on the year
#             rgba_color = cmap(norm(year))  # Get RGBA color from colormap
#             hex_color = mcolors.to_hex(rgba_color)  # Convert to hex for folium
            # Get color for the current year
            color = colormap(year)  # Use color from colormap based on year


            # Use an HTML popup to display the photo and date
            popup_html = f'<img src="{photo_folder}/{filename}" width="300px"><br>Date taken: {date_time}'
            folium.Marker(
                location=gps_coords,
                popup=folium.Popup(popup_html, max_width=300),
                icon=folium.Icon(color="white", icon="info-sign", icon_color=color)
            ).add_to(map_)
# Add the color bar to the map
colormap.add_to(map_)

# Save the map to an HTML file
map_.save("photo_locations_map.html")
print("Map saved as photo_locations_map.html")
```
